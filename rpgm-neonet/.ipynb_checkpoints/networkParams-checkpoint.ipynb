```python
import rpgm_algo as rpg
import numpy as np
import scipy.optimize as so
import awesomeplot.core as plt
%matplotlib notebook
```

```python
def get_network(x, coords):   
    g = rpg.RpgAlgorithm()
    g.debug = False
    g.set_params(n=int(len(coords)), n0=int(x[0]), p=x[1], q=x[2], r=x[3], s=x[4])
    g.setup_locations(sampling="data", locations=coords)
    g.initialise()
    g.grow()
    return g

def objective_function(x, coords):
    g = get_network(x, coords)
    md = np.mean(np.sum(g.adjacency.todense(), axis=1))
    return (np.mean(md) - 2.7)**2 
```

```python
n = 100
x = [1, .5, .5, 1./3., .5]

h = rpg.RpgAlgorithm()
coords = np.zeros([n, 2])
for i in xrange(n):
    coords[i] = h._get_coords(sampling="uniform", centre=[.5, .5], boundaries=[0.5, 0.5])

bounds = [
    [2, n - 1],
    [0., 1.],
    [0., 1.],
    [0., 10.],
    [0., 1.]
]

print objective_function(x, coords)
```

```python
sol = so.minimize(objective_function, x, args=(coords,), method="SLSQP", bounds=bounds,  
                  options={"disp":True, "maxiter":50})
print sol.message
print "Success:", sol.success
print sol.x
print objective_function(x, coords)
```

```python
sol = so.differential_evolution(objective_function, bounds, args=(coords,), maxiter=50, tol=1.)

print sol.message, sol.nfev, "func evals", sol.nit, "iterations"
print "Success:", sol.success
print sol.x
print objective_function(x, coords)
```

```python
g = get_network(sol.x, coords)
print np.mean(np.sum(g.adjacency.todense(), axis=1))

canvas = plt.Plot()
layout = g.locations
fig = canvas.add_network(g.adjacency.todense(), sym=True, vertex_labels=range(g.n),
                   styles={"layout":layout})
fig.axes[0].axis("off")
```
